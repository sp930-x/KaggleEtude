## Q3. Where should I learn data science?

```python
mcq['LearningPlatformSelect'] = mcq['LearningPlatformSelect'].astype('str').apply(lambda x2: x2.split(','))
s2 = mcq.apply(
 lambda x2: pd.Series(x2['LearningPlatformSelect']),
 axis=1).stack().reset_index(level=1, drop=True)
s2.name = 'platform'
```

```python
plt.figure(figsize=(6,8))
data = s2[s2 != 'nan'].value_counts().head(15)
sns.barplot(y=data.index, x=data)
```

The result shows that Kaggle is the most popular learning platform. However, since the survey was conducted by Kaggle there should be some bias in the answers.

```python
# 설문내용과 누구에게 물어봤는지를 찾아봄
question = pd.read_csv('schema.csv')
qc = question.loc[question['Column'].str.contains('LearningCategory')]
print(qc.shape)
qc
```

```python
use_features = [x for x in mcq.columns if x.find('LearningPlatformUsefulness')!=-1]
```

```python
#Relationship between the learning platform and the usefulness
fdf = {}
for feature in use_features:
    a = mcq[feature].value_counts()
    a= a/a.sum() #to express in percentage
    fdf[feature[len('LearningPlatformUsefulness'):]] = a
    
fdf = pd.DataFrame(fdf).transpose().sort_values(
    'Very useful', ascending=False)

#Draw plot
plt.figure(figsize=(10,10))
sns.heatmap(
    fdf.sort_values(
    "Very useful", ascending=False), annot =True)

```

```python
# 유용함의 정도를 플랫폼별로 그룹화해서 본다.
fdf.plot(kind='bar', figsize=(20,8),
 title="Usefullness of Learning Platforms")
```

```python
cat_features = [x for x in mcq.columns if x.find('LearningCategory') != -1]
cat_features
```

```python
cdf = {}
for feature in cat_features:
    cdf[feature[len('LearningCategory'):]] = mcq[feature].mean()
cdf = pd.Series(cdf)
cdf
```

```python
plt.pie(cdf, labels=cdf.index,
       autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title("Contribution of each Platform to Learning")
plt.show()
```

## Q5. How much time do you use for data science?

```python
plt.figure(figsize=(6, 8))
sns.countplot(y='TimeSpentStudying',
 data=mcq,
 hue='EmploymentStatus'
 ).legend(loc='center left',
 bbox_to_anchor=(1, 0.5))
```

```python
figure, (ax1, ax2) = plt.subplots(ncols=2)

full_time = mcq.loc[(mcq['EmploymentStatus'] == 'Employed full-time')]
looking_for_job = mcq.loc[(
 mcq['EmploymentStatus'] == 'Not employed, but looking for work')]

figure.set_size_inches(12,5)
sns.countplot(x='TimeSpentStudying',
 data = full_time,
 hue='EmploymentStatus', ax=ax1
 ).legend(loc='center right',
 bbox_to_anchor=(1, 0.5))

sns.countplot(x='TimeSpentStudying',
 data = looking_for_job,
 hue='EmploymentStatus', ax=ax2
 ).legend(loc='center right',
 bbox_to_anchor=(1, 0.5))
```

###### Definitely the job seekers tend to spend more time on the study.
