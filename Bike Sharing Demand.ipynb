## 탐색적데이터분석

#### Data Fields
* datetime - hourly date + timestamp  
* season -  1 = spring, 2 = summer, 3 = fall, 4 = winter 
* holiday - whether the day is considered a holiday
* workingday - whether the day is neither a weekend nor holiday
* weather   
    1. Clear, Few clouds, Partly cloudy, Partly cloudy
    2. Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    3. Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    4. Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
* temp - temperature in Celsius
* atemp - "feels like" temperature in Celsius
* humidity - relative humidity
* windspeed - wind speed
* casual - number of non-registered user rentals initiated
* registered - number of registered user rentals initiated
* count - number of total rentals

```python
train["year"] = train["datetime"].dt.year
train["month"] = train["datetime"].dt.month
train["day"] = train["datetime"].dt.day
train["hour"] = train["datetime"].dt.hour
train["minute"] = train["datetime"].dt.minute
train["second"] = train["datetime"].dt.second

train.shape
```

```python
train.head()
```

```python
figure,((ax1,ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(nrows=2, ncols=3)
figure.set_size_inches(18,8)

sns.barplot(data=train, x="year", y="count", ax=ax1)
ax1.set(ylabel='Count', title="Yearly counts")

sns.barplot(data=train, x="month", y="count", ax=ax2)
ax2.set(xlabel='month', title="Monthly counts")

sns.barplot(data=train, x="day", y="count", ax=ax3)
ax3.set(xlabel='day', title="Daily counts")

sns.barplot(data=train, x="hour", y="count", ax=ax4)
ax2.set(xlabel='hour', title="Hourly counts")

sns.barplot(data=train, x="minute", y="count", ax=ax5)
sns.barplot(data=train, x="minute", y="count", ax=ax6)
```

* For the yearly counts, people used bike more in 2012 than 2011.
* From the monthly perspective, June is the best but counts are high also from July to October. January is the slowest period.
* Daily data lack after 20th, so this data can't be used.
* Minutes and second basis data are useless since its input data are all 0.
* On Hourly basis, tit is high on 17, 18, 19, and 8 hour. This trend indicates that many demands are caused due to the commute.

```python
corrMatt = train[["temp","atemp","casual","registered","humidity","windspeed","count"]]
corrMatt = corrMatt.corr()
print(corrMatt)

mask = np.array(corrMatt)
mask[np.tril_indices_from(mask)]=False
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20,10)
sns.heatmap(corrMatt, mask=mask,vmax=.8, square=True,annot=True)
```

* temp, humidty, windspeed has a few relation.
* 'Registered' seems to have the highest relations with 'count'. However test data doesn't have it.
* atemp and temp has strong relation but this will cause multicollinearity, which means not appropriate to use.

```python
def concatenate_year_month(datetime):
    return "{0}-{1}".format(datetime.year, datetime.month)

train["year_month"] =train["datetime"].apply(concatenate_year_month)
print(train.shape)
```

```python
train[["datetime","year_month"]].head()
```

```python
fig,(ax1,ax2) = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(18,4)

sns.barplot(data=train, x="year", y="count",ax=ax1)
sns.barplot(data=train, x="month", y="count",ax=ax2)

fig, ax3 = plt.subplots(nrows =1, ncols=1)
fig.set_size_inches(18,4)

sns.barplot(data=train, x="year_month", y="count", ax=ax3)
```

#### Getting rid of outliers

```python
#get rid of outliers over 3 sigmas
trainWithoutOutliers = train[np.abs(train["count"] - train["count"].mean())<=(3*train["count"].std())]

print(train.shape)
print(trainWithoutOutliers.shape)
```

```python
figure, axes = plt.subplots(ncols=2, nrows=2)
figure.set_size_inches(12, 10)

sns.distplot(train["count"], ax=axes[0][0])
stats.probplot(train["count"], dist='norm', fit=True, plot=axes[0][1])
sns.distplot(np.log(trainWithoutOutliers["count"]), ax=axes[1][0])
stats.probplot(np.log1p(trainWithoutOutliers["count"]), dist='norm', fit=True, plot=axes[1][1])
```
